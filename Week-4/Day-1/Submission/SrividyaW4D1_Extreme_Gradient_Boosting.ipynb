# Perform Feature Engineering on our dataset and will show it.

```python
original_features= list(df.columns)
print(original_features)
```

```python
dataset_array = np.array(df)
```

```python
print(dataset_array)
```

```python
features_indicies = []
attributes = ['url', 'domain', 'file', 'params']
```

```python
new_dataset = {}
```

```python
for index, name in enumerate(original_features): #index represents index of current and name represents the name in the current feature
  if 'qty' in name and name.split('_')[-1] in attributes: # checking 'qty' in the substring
    features_indicies.append([index, name.split('_')[-1]]) # last part is from attributes
  else:
    new_dataset[name] = dataset_array[:,index] # it assigns the columns that dont meet the condition to the new dictionary
```

```python
for index, attribute in features_indicies: #this line iterates over the features_indices list
  if attribute == 'domain':
    if f'qty_char_{attribute}' not in new_dataset.keys():
      new_dataset[f'qty_char_{attribute}'] = np.zeros(rows) #it creates new key in the new_dataset with the name  ex:qty_char{attribut}

    new_dataset[f'qty_char_{attribute}']+= dataset_array[:, index]
```

```python
df1 = pd.DataFrame(new_dataset).astype(int) #to convert  data in interger datatypes
```

```python
df1[df1<-1] = -1 # if the condition is true , i want to change the element to -1
df1
```

```python
# summary statastics of our dataset

print(df1.describe())
```

```python
# setting up the data for modelling

setup(data = df1, target = "phishing") # target variable is phishing which we want to build model
```

Its Target is Binary because it has 2 classes 0 and 1

```python
# comparing and select the best data model

best_model = compare_models()
```

```python
tuned_model = tune_model(best_model,n_iter=1, optimize = 'F1')
```

```python
predictions = predict_model(tuned_model, data = df1)
```
